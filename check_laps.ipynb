```python
def add_laps(s1, s2):
    new_lap = s2.copy()
    new_lap.start_pos_lat  = s1.start_pos_lat
    new_lap.start_pos_long = s1.start_pos_long
    new_lap.lap_time = s1.lap_time + s2.lap_time
    new_lap.lap_distance = s1.lap_distance + s2.lap_distance
    new_lap.avg_speed = new_lap.lap_distance / new_lap.lap_time
    new_lap.start_x = s1.start_x
    new_lap.start_y = s1.start_y

    return new_lap

# add_laps(laps.iloc[9], laps.iloc[10])
```
